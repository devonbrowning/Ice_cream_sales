```python
df = pd.read_csv('monthlysales.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['Total Units'], color='b')
plt.title('Ice Cream Profit by Month')
plt.xlabel('Month')
plt.ylabel('Total Profit')
```

```python
plt.figure(figsize=(9, 6))

for flavor in ['Vanilla', 'Chocolate', 'Strawberry', 'Neapolitan', 'Mango', 'Lime']:
    plt.plot(df['Month'], df[flavor], label=flavor)

plt.title('Ice Cream Sales by Flavor Over Time')
plt.xlabel('Month')
plt.ylabel('Ice Cream Profit')
plt.legend()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Month'], df['Total Profit'], color='c')
plt.title('Total Profit per Month')
plt.xlabel('Month')
plt.ylabel('Total Profit ($)')
plt.grid(axis='y')
```
